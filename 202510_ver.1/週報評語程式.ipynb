```python
appetite = 'good' #good,bad
nutrition = 'perfect' #over, less, perfect
poop = 'normal' #normal, soft

if appetite == 'good':
    if nutrition == 'perfect':
        if poop == 'normal':
            print('健康寶貝for sure')
        elif poop == 'soft':
            print('要注意寶的便便')
    elif nutrition == 'less' or nutrition == 'over':
        if poop == 'normal':
            print('營養要注意唷')
        elif poop == 'soft':
            print('注意營養及便便')
elif appetite == 'bad':
    if poop == 'soft':
        print('寶貝吃不下又烙賽了！紅色警戒')
    elif nutrition == 'less' or nutrition == 'over':
        print('寶貝營養攝取有狀況，請注意')
    elif nutrition == 'perfect':
        print('寶貝最近吃不下')
else:
    print('寶貝最近吃不下')
```
