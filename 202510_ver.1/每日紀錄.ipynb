```python
daily_tracker = []

while True:
    date = input('今天日期yyyy-mm-dd(結束輸入請輸入send):')
    if date == ('send'):
        break
    food_type = input('類別:')
    food_brand = input('品牌:')
    food_name = input('品項:')
    food_flavor = input('口味:')
    food_weight = input('克數:')
    finished_or_not = input('寶貝有吃光嗎？YES / NO:')
    poop = input('便便如何？沒便便 / 軟便便 / 完美便:')

    tracker = {'今天日期':date, '類別':food_type, '品牌':food_brand, '品項':food_name, '口味':food_flavor, '克數':food_weight, '寶貝有吃光嗎？': finished_or_not, '便便如何？':poop}
    daily_tracker.append(tracker)

    print("\n每日紀錄表")
    print('今天日期｜類別｜品牌｜品項｜口味｜克數｜寶貝有吃光嗎？｜便便如何？')
    for a in daily_tracker:
        print(f"{a['今天日期']}｜{a['類別']}｜{a['品牌']}｜{a['品項']}｜{a['口味']}｜{a['克數']}｜{a['寶貝有吃光嗎？']}｜{a['便便如何？']}")
```
